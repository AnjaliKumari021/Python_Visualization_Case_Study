### 1. Compare Sales by region for 2016 with 2015 using bar chart


```python
Region_Sales = Sales.groupby(by='Region')[['Sales2015','Sales2016']].sum()
```

```python
Region_Sales
```

```python
# plot the bar graph

plt.figure(figsize=(10,8))
Region_Sales.plot(kind='bar')
plt.ylabel('Sales')
plt.title('Sales by region for 2016 and 2015')
plt.xticks(rotation=0)
plt.show()
```

*From the above graph we can observe that the sales are higher in 2016 as compared to 2015 for all the regions. Of all the regions, East region is contribution maximum in sales.*

### 6. Create a new column Qtr using numpy.where() or any suitable utility in the imported dataset. The Quarters are based on months and defined as -
*Jan - Mar : Q1* 

*Apr - Jun : Q2* 

*Jul - Sep : Q3* 

*Oct - Dec : Q4*


```python
Sales.head(3)
```

```python
# Using Numpy.where for adding qtr column
Sales['Qtr'] = np.where(Sales['Month'].isin(['Jan','Feb','Mar']),'Q1',
                               np.where(Sales['Month'].isin(['Apr','May','Jun']),'Q2',
                                   np.where(Sales['Month'].isin(['Jul','Aug','Sep']),'Q3','Q4')))
```

```python
# qtr column added in the dataframe
Sales.head(3)
```

### 7. Compare Qtr wise sales in 2015 and 2016 in a bar plot


```python
Qtr_Sales = Sales.groupby(by='Qtr')[['Sales2015','Sales2016']].sum()
```

```python
Qtr_Sales
```

```python
# plotting bar graph for comparing quarter wise sales

Qtr_Sales.plot(kind='bar')
plt.xticks(rotation=0)
plt.title('Quarter wise Sales in 2015 and 2016')
plt.ylabel('Sales')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

*From the above graph we can see that in Q3 the sales are high in 2016 as compared to 2015*

### 8. Determine the composition of Qtr wise sales in and 2016 with regards to all the Tiers in a pie chart.
#### (Draw 4 pie charts representing a Quarter for each Tier)

```python
Sales.head(3)
```

```python
Qtr_tier_pie = Sales.pivot_table(index='Qtr',columns='Tier', values='Sales2016',aggfunc='sum')
```

```python
Qtr_tier_pie
```

```python
# updating the negative out value with 0.0
Qtr_tier_pie.iloc[2,3] = 0.0
```

```python
# updated pivot table
Qtr_tier_pie
```
